```python
import pandas as pd
```

```python
base = pd.read_csv('FIFA World Cup All Goals 1930-2022.csv', encoding='ISO-8859-1')
```

```python
base = base.rename(columns={'key_id':'Gol de numero','goal_id':'ID Gol','tournament_id':'Ano','tournament_name':'Nome do torneio','match_id':'ID Jogo','match_name':'Nome do jogo','match_date':'Data do jogo','stage_name':'Estágio','group_name':'Nome do grupo','team_id':'ID do time','shirt_number':'Numero da camisa','player_team_id':'ID do 444','player_team_name':'Nome do 444','player_team_code':'Abreviação','minute_label':'Minuto regulamentado','minute_regulation':'Minuto marcado','minute_stoppage':'Acrescimos','match_period':'Periodo marcado','own_goal':'Gol contra','penalty':'Penalti','given_name':'Nome do jogador','team_name':'Nome do time','family_name':'Sobrenome'})
```

```python
base['Jogador'] = base['Nome do jogador'].str.cat(base['Sobrenome'],sep=' ')
```

```python
base=base.drop(['Minuto regulamentado','Nome do torneio','ID do 444','player_id','away_team','home_team','team_code','Data do jogo','ID Gol','Abreviação','Nome do 444','Sobrenome','Nome do jogador'],axis=1)
```

```python
base['Ano'] = base['Ano'].str.replace('WC-','')
base['ID do time'] = base['ID do time'].str.replace('T-','')
base['ID Jogo'] = base['ID Jogo'].str.replace('M-','').str.lstrip('0')
base['ID Jogo'] = base['ID Jogo'].str.replace('-','')
```

```python
base
```

```python
qtd_gols = base['Gol de numero'].value_counts().sum()
```

```python
base[base['Periodo marcado']=='second half, stoppage time'].sort_values('Acrescimos',ascending = False)
```

```python
media_gol = base['Gol de numero'].value_counts().sum() / base['ID Jogo'].nunique()
```

```python
media_gol
```

```python
gol_ano=base.groupby(['Ano','ID Jogo'])['ID Jogo'].nunique().groupby('Ano').sum()
```

```python
jogos_ano=base.groupby(['Ano','Gol de numero'])['Gol de numero'].nunique().groupby('Ano').sum()
```

```python
import numpy as np
media_gols = np.array(jogos_ano)/np.array(gol_ano)
media_gols = np.round(media_gols,1)
media_gols
```

```python
dic= {}
a=0
for year in base['Ano'].unique():
    dic[int(year)]=f'A média de gols foi de: {media_gols[a]}'
    a+=1
    
    
dic
```

```python
time_2=time_2['Gol de numero'].value_counts().sum()
time_1=time_1['Gol de numero'].value_counts().sum()
total = time_1 + time_2
porce_1 = round((time_1/total)*100,2)
porce_2 =  round((time_2/total)*100,2)
dic_porce ={}
dic_porce['Primeiro tempo'] = porce_1
dic_porce['Segundo tempo'] = porce_2
dic_porce
```

```python
gol_fase_grupo = base[base['Estágio']=='group stage'].value_counts().sum()
gol_fase_mata = base[base['Estágio']!='group stage'].value_counts().sum()
```

```python
total1 = gol_fase_grupo + gol_fase_mata
gol_g = round((gol_fase_grupo/total1)*100,2)
gol_m =  round((gol_fase_mata/total1)*100,2)
dic_gol ={}
dic_gol['Gol em fase de grupos'] = gol_g
dic_gol['Gol em mata-mata'] = gol_m
dic_gol
```
